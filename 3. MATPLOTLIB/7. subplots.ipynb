# subplots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4,5]
y = [20,45,67,44,50]
y1 = np.arange(10,51,10)
y2 = np.arange(5,26,5)
y3 = np.arange(50,9,-10)
```

```python
plt.figure(figsize = (10,5))
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.scatter(x,y1)

plt.subplot(2,2,3)
plt.barh(x,y2)

plt.subplot(2,2,4)
plt.bar(x,y3)

plt.savefig("subplot.png")
plt.show()
```
