```python
abc.plot(x='Date', y='BTC-USD Price',
         label='Bitcoin Price',
         linewidth =3, figsize=(14,6))
plt.ylabel ('Price[$]')
plt.xlabel ('Date')
plt.title ('My First Data Visualization')
plt.legend(loc ='upper left')
plt.grid
```

```python
abc.plot(x='Date', y='BTC-USD Price',
         label='Bitcoin Price',
         linewidth =3, figsize=(14,6))
         
plt.ylabel ('Price[$]',color='Black',fontsize=11,fontname='ALGERIAN')
plt.xlabel ('Date', fontsize = 17 , fontname='ALGERIAN')
plt.title ('My First Data Visualization',color='Red',fontsize=17, fontname='Snap ITC')
plt.legend(loc ='upper left')
plt.grid
```

```python
plt.plot(x='Date',y='ETH-USD Price',label='Bitcoin',linewidth=2.5,figsize=(10,5),color='b')
plt.ylabel('Price[$]',color='Black',fontsize=11,fontname="ALGERIAN")
plt.xlabel('Date',fontsize=17,fontname='ALGERIAN')
plt.title('My First Data Visualization Exercise',color='Black',fontsize=17,fontname='Snap ITC')
plt.legend(loc="upper left")
plt.grid(linewidth=2)
```
