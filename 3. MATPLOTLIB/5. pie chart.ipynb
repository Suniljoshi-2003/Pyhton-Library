# pie chart

```python
import matplotlib.pyplot as plt
```

```python
x = ["Apple","Samsung","Lenovo","HP","Asus","Dell" ]
y = [450,500,230,700,800,300]
c = ["green", '#5bc9b1',"red","#924e99","orange","hotpink"]
ex = [0,0,0,0.1,0,0.2]

plt.pie(y, labels = x, colors = c, explode = ex, shadow = True, autopct = "%0.2f%%", startangle = 90 )
plt.show()
```
