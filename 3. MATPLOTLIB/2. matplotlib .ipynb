# Matplotlib
matplotlib.pyplot is collection of cammand style function that make matplotlib work like MATLAB.

Matplotlib is a Python 2D  plotting library which produces publication-quality figires in a variety of hardcopy formats and interactive enviroments across platfoms.

```python
#  imorting matplotlib
import matplotlib.pyplot as plt 
```

```python
plt.plot([2,4,6,4,8]) # taking random number
plt.ylabel("Numbers")
plt.xlabel("Indies")
plt.title("MyPlot")
plt.show()
```

```python
# [1,2,3,4]=x & [1,4,9,16]=y
plt.plot([1,2,3,4],[1,4,9,16])
plt.ylabel("squares")
plt.xlabel("numbers")
plt.grid() # grid on
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.grid()
plt.show()
```

```python
# now can you observe same changes with respect to above one
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.grid()
plt.show()
```

```python
#  for your reference we use numpy for matematical operations
import numpy as np
t = np.arange(0.,5.,0.2)

#  blue dashes , red squares and green traingles
plt.plot(t, t**2, 'b--', label ='^2')  # 'rs', 'g^'  # blue dash line.
plt.plot(t, t**2.2,'rs',label='^2.5')  # red square (rs)
plt.plot(t, t**2.5, 'g^',label = '^2.5') # green traingle (g^)
plt.grid()
plt.legend() # add legend based on line labels
plt.show()
```

- countrolling line properties

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y, linewidth = 5.0) # change the value of linewidth=5.0 to any other value to observe the changes
plt.grid()
plt.show()
```

# Barchart

```python
x= [2,4,6,8,10]
y= [6,7,8,2,4]

x2 = [1,3,5,9,11]
y2 = [7,8,2,4,2]

plt.bar(x,y, label='Bar1')
plt.bar(x2,y2, label = 'Bar2')

plt.xlabel('x')
plt.ylabel('y')

plt.title('Intresting graph\n check it out')

plt.legend()

plt.show()
```
