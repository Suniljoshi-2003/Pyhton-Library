```python
x=pd.DataFrame({'Name':['Python','Java','Python','HTML','C++'],
                'Salary':[30,30,22,22,21],
                'Cource':['DA','DS','DE','DS','DA']})
x
```

```python
#  groupby --> Aggrigate at the end
x.groupby('Cource')['Salary'].sum()
#  unordered 
```

```python
#  unordered 
z = x.groupby('Cource') ['Salary'] .sum()
z.sort_values(ascending=False,inplace=True) # desc to Asc
z
```
